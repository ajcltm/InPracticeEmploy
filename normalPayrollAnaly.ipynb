```python
from pathlib import Path

import pandas as pd
from datetime import datetime

from normalPayroll import NormalPayroll
```

```python
normPay = NormalPayroll()
```

```python
df1 = normPay.loadData()
```

```python
df1['직책'] = df1['직책'].replace({'본부장(직무대행)':'본부장'})
```

```python
df2 = normPay.mergeSudangDf(df1)
df2
```

```python
df3 = df2
df3['연도'] = df2['적용년월'].apply(lambda x: x.strftime('%Y'))
df3 = df3.groupby(['사원', '연도']).last()
```
