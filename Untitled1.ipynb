```python
import pandas as pd
```

```python
path = 'C:/Users/user/Desktop/직원 현황 테스트/210616_사원명부.csv'
file = pd.read_csv(path, encoding='utf-8')
df = file[['사원', '직급']]
df
```

```python
levels = ['운영국장', '본부장', '1급', '2급']
df_n = df[df['직급'].isin(levels)]
df_n = df_n.sort_values(by='직급')
df_n = df_n.reset_index(drop=True)
df_n
```

```python
df_sum=df_n.groupby('직급').count()
df_sum
```
